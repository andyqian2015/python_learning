```python
import numpy as np
import matplotlib.pyplot as plt
# ref
# https://blog.csdn.net/hurry0808/article/details/78013691
```

```python
x = np.arange(0, 2 * np.pi, 0.001)
y = np.sin(2 * np.pi *x)
```

```python
plt.plot(x, y)
```

```python
plt.clf()
plt.plot(x, y)
l = plt.axhline(linewidth=1, color='black')
l = plt.axvline(linewidth=1, color='black')
plt.show()
```
