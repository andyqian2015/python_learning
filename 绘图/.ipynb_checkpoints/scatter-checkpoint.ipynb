```python
import matplotlib.pyplot as plt
import numpy as np
#　ref
# http://www.sofasofa.io/forum_main_post.php?postid=1007596
# https://blog.csdn.net/u010852680/article/details/77770097
# https://www.cnblogs.com/sench/p/9522627.html
```

```python
x = np.random.randint(0, 10, 3)
y = np.random.randint(0, 10, 3)
```

```python
x
```

```python
y
```

```python
# 使用同样的颜色绘制点，默认为圆点
plt.scatter(x, y, c = 'r', cmap='BuGn')
```

```python
# 如果颜色是一个列表的数组，则对应点的长度要和颜色的长度一致
plt.scatter(x, y, c = ['r', 'g', 'b'], cmap='BuGn')
```

```python
# 这个代表产生1,2的均匀分布,产生的数据为10个
np.random.uniform(1, 2, 10)
```

```python
plt.scatter(x, y, c = 'r', cmap='BuGn')
```

```python
# 这种颜色方式，出现显示点的问题，可以之后再查询下是什么问题
plt.scatter(x, y, c = np.random.uniform(2, 3, 3), cmap='BuGn')
```

```python
#　更改绘制的图像的形状,通过marker参数
# marker参数使用特定的符号*
plt.scatter(x, y, c = 'r', cmap='BuGn', marker="*")
```

```python
# 绘制三点
plt.scatter(x, y, c = 'r', cmap='BuGn', marker="$...$", s =500)
```

```python
# 参数化绘制
#　(numsides, style, angle)
#　numsides为边的数目，style为０~３，angle为旋转角度
plt.scatter(x, y, c = 'r', cmap='BuGn', marker=(4, 0, 0), s =500)
```

```python
plt.scatter(x, y, c = 'r', cmap='BuGn', marker=(4, 1, 0), s =500)
```

```python
plt.scatter(x, y, c = 'r', cmap='BuGn', marker=(4, 2, 0), s =500)
```

```python
# 当stype为３时角度和边自动忽略
plt.scatter(x, y, c = 'r', cmap='BuGn', marker=(4, 3, 0), s =500)
```

```python
plt.scatter(x, y, c = 'r', cmap='BuGn', marker=(4, 0, 45), s =500)
```

```python
plt.scatter(x, y, c = 'r', cmap='BuGn', marker=(4, 2, 45), s =500)
```

```python
# 绘制形状的大小，通过绘制中的参数s设置
plt.scatter(x, y, c='r', s= 100)
```

```python
# 设置点的透明度:alpha
plt.scatter(x, y, c='r', s= 100, alpha=0.5)
```

```python
#　颜色值可以用数表示
colors = np.random.rand(3) # 产生[0, 1]范围内的颜色的随机值
```

```python
plt.scatter(x, y, c=colors, s= 100, alpha=0.5)
```

```python
# 为散点图设置图例
'''
    ===============   =============
        Location String   Location Code
        ===============   =============
        'best'            0
        'upper right'     1
        'upper left'      2
        'lower left'      3
        'lower right'     4
        'right'           5
        'center left'     6
        'center right'    7
        'lower center'    8
        'upper center'    9
        'center'          10
        ===============   =============
'''

plt.scatter(x, y, c=colors, s= 100, alpha=0.5, label="circle")
plt.legend(loc="best")
```
